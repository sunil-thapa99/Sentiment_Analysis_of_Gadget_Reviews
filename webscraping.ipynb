```python
blog1 = 'https://www.tomsguide.com/'
blog2 = 'https://www.digit.in/reviews/mobile-phones/apple-iphone-14-pro-max-review-263191.html'
blog3 = 'https://mashable.com/'
```

```python
import requests
from bs4 import BeautifulSoup
```

### For blog-1

```python
additional = 'reviews/iphone-14-pro-max'
url1 = blog1 + additional
response = requests.get(url1)
```

```python
# create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
# Find the first <div> tag with an id attribute of "my-div"
div_tag = soup.find('div', id="article-body")

# Access the text content of the <div> tag
content = div_tag.text
content
```

```python
# converting users input into query
users_input = 'iphone 14 pro'
users_query = users_input.split(' ')
final_user_query = ''
for index, i in enumerate(users_query, 0):
    if index !=0 :
        final_user_query += '+'
        final_user_query += i
    else:
        final_user_query += i
final_user_query
```

```python
# scraping the searching page with users input
search_query = 'search?searchTerm='+ final_user_query
url1 = blog1 + search_query
response = requests.get(url1)

# create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
# Getting the list of search results
div_tag = soup.find('div', class_='listingResults')
div_tag
```

```python
# Listing the search results 
search_results = div_tag.find_all('div', class_='listingResult')
```

```python
len(search_results)
```

```python
# Accessing and Printing the article title and its link from search results
for each_blog in search_results:
    print("Title:", each_blog.find('article').text)
    print("Link:", each_blog.find('a', href=True)['href'])
```

```python
# extracting the bolg contents 
article_text = []
count = 1

# temp_response = requests.get(tmp_url)
for each_blog in search_results:
    name = "article"+str(count)
    count += 1
    try:
        # getting blog url
        tmp_url = str(each_blog.find('a', href=True)['href'])

        # getting the url
        temp_response = requests.get(tmp_url)

        # creating soup object
        temp_soup = BeautifulSoup(temp_response.content, "html.parser")
        print("temp_soup", type(temp_soup))

        # getting the text from the article with class => news-article 
        article_div = temp_soup.find('article', class_='news-article')    
        article_header = article_div.find('h1').text
        article_body_text = article_div.find('section').text
        article_text.append([name, article_header, article_body_text])
        
    except:
        continue
    
```

## Converting the data to dataframe

```python
df = pd.DataFrame(article_text, columns=['article', 'header', 'text'])
df
```

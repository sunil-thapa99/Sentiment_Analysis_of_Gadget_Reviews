```python
import re
import json
import requests
from bs4 import BeautifulSoup
```

```python

URL = "https://www.youtube.com/@mkbhd/videos"
soup = BeautifulSoup(requests.get(URL).content, "html.parser")
```

```python
data = re.search(r"var ytInitialData = ({.*});", str(soup)).group(1)
```

```python
json_data = json.loads(data)
```

```python
contents = json_data['contents']['twoColumnBrowseResultsRenderer']['tabs'][1]['tabRenderer']['content']['richGridRenderer']['contents']
```

```python
count = 1
for content in contents:
  vid_id = content['richItemRenderer']['content']['videoRenderer']['videoId']
  title = content['richItemRenderer']['content']['videoRenderer']['title']['runs'][0]['text']
  print(f'vid_id: {vid_id}, title: {title}')
  count+=1
  if count > 20:
    break
```
